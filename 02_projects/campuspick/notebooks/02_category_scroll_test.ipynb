```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
# 1) CSVì—ì„œ ì¹´í…Œê³ ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
category_csv = "../data/contest_categories.csv"
df_categories = pd.read_csv(category_csv)
print(f"[INFO] Loaded {len(df_categories)} categories from CSV")
df_categories.head()

# 2) Selenium ë“œë¼ì´ë²„ ì‹¤í–‰
driver = webdriver.Chrome()   
print("[INFO] Chrome driver started")
```

```python
# 3) ìŠ¤í¬ë¡¤ : ì²˜ìŒ 2~3ê°œ ì¹´í…Œê³ ë¦¬ë§Œ í…ŒìŠ¤íŠ¸
for index, row in df_categories.head(3).iterrows():
    cat_id = row["category_id"]
    cat_name = row["category_name"]
    url = f"https://www.campuspick.com/contest?category={cat_id}"
    
    driver.get(url)
    print(f"[INFO] Loaded category '{cat_name}' at {url}")
    time.sleep(2)  # JS ë Œë”ë§ ëŒ€ê¸°

    # ìŠ¤í¬ë¡¤ í…ŒìŠ¤íŠ¸ (ë™ì  ë¡œë”© í™•ì¸)
    scroll_pause = 1
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    for _ in range(3):  # 3ë²ˆ ì •ë„ ìŠ¤í¬ë¡¤
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break  # ë” ì´ìƒ ë¡œë”© ì—†ìœ¼ë©´ ì¢…ë£Œ
        last_height = new_height

    print(f"[INFO] Finished scrolling category '{cat_name}'\n")
```

```python
# ë¸Œë¼ìš°ì € ì¢…ë£Œ
driver.quit()
print("[INFO] Chrome driver closed")
```
