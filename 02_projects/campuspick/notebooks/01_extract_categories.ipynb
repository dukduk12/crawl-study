```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
# 1) Selenium ë“œë¼ì´ë²„ ì‹¤í–‰ (Chrome) ë° ì‚¬ì´íŠ¸ ì ‘ì†
driver = webdriver.Chrome()   
print("[INFO] Chrome driver started")

url = "https://www.campuspick.com/contest"
driver.get(url)
time.sleep(2)  # JS ë Œë”ë§ ëŒ€ê¸°
print(f"[INFO] Page loaded: {url}")
```

```python
# 2) ì¶”ì¶œ : section.tab ì•ˆì˜ wrap í´ë˜ìŠ¤ div ì•ˆ a[data-category] ì„ íƒ
category_elements = driver.find_elements(By.CSS_SELECTOR, "section.tab .wrap a[data-category]")

categories = []
for el in category_elements:
    cat_id = el.get_attribute("data-category")
    cat_name = el.text.strip()
    categories.append({
        "category_id": cat_id,
        "category_name": cat_name
    })
    print(f"[DEBUG] Extracted category: ID={cat_id}, Name={cat_name}")
```

```python
# 3) DataFrame ìƒì„±
df_categories = pd.DataFrame(categories)
print("[INFO] Categories DataFrame created")
print(df_categories)
```

```python
# 4) CSV ì €ì¥
df_categories.to_csv("../data/contest_categories.csv", index=False)
print("[INFO] Categories saved to ../data/contest_categories.csv")
```
